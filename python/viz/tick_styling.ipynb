Once you've got the correct strings printing along the axes, with as many (or few) ticks as you want, and at the right spacing, there are a lot of options for the way you can adjust the *style* of the rendering.

Let's start with a simple graph.

```python
%pylab inline

x = y = np.linspace(0, 10)

fig, ax = plt.subplots()
ax.plot(x, y)
```

## Changing Appearance

The `tick_params()` method for an `Axis` object gives you access to a [whole bunch of options](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.tick_params.html) for formatting your ticks and their labels, including:

If the tick faces in or out

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               direction='in')
```

How much the label is rotated

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               rotation=90)
```

How thick a tick is

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               width=10)
```

The space between label and tick

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               pad=20)
```

The color of the ticks

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               color='orange',
               width=5)
```

## Hiding Them Altogether

We can also control whether or not the tick shows up

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               bottom=False)
```

Or the label

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x',
               labelbottom=False)
```

Or all of it

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='both',
               labelbottom=False,
               bottom=False,
               labelleft=False,
               left=False)
```
